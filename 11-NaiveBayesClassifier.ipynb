```python
#here we can actually implement encoder before we do train-test split.
#because we are encoding y not X. there is no risk of data leakage since model does not know 
#anything about y when it is training in the first place
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
df['Species'] = label_encoder.fit_transform(df['Species'])
```

```python
df.head()
```

```python
df.tail()
```

```python
df['Species'].value_counts()
```

```python
X= df.drop(["Species"],axis =1)
y= df["Species"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25, random_state=15)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
y_pred=gnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print("confusion matrix: \n ", confusion_matrix(y_pred,y_test))
print("accuracy score: ", accuracy_score(y_pred,y_test))
print("classification report: ", classification_report(y_pred,y_test))
```
